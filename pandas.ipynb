## Lendo arquivo csv e Alterando o separador --> sep = '...'

```python
df = pd.read_csv('vendas.csv',sep=';')
df
```

## criação de coluna

```python
df['Faturamento']= df['quantidade_vendida'] * df['preco_unitario']
df
```

```python
faturamento_total= df['Faturamento'].sum()
faturamento_medio = df ['Faturamento'].mean()
print(f'Total {faturamento_total}\nMédio {faturamento_medio}')
```

## Criando 'novo' csv

```python
df.to_csv('vendaspython.csv',sep=';', index= False)
```
