```python
def adicionar_aluno(lista):
    aluno = input('Digite o nome do Aluno(a): ')
    lista.append(aluno)
    print(f'Aluno(a) {aluno} cadastrado!')



    


def tratar_altura(funcao):
    while True:
        try:
            valor = float(input(funcao))
            return valor
        except ValueError: 
            print('Digite um número válido')
    

def tratar_peso(peso):
    while True:
        try:
            a1 = float(input(peso))
            return a1
        except ValueError: 
            print('Digite um número válido')


def tratar_idade(codigo):
    while True:
        try: 
         valor = int(input(codigo))
         return valor
        except ValueError:
            print('Digite um número válido')

def calculo_imc(peso,altura):
    resultado = peso / altura * altura
    return resultado


```

```python
dicionario = {'nome': [],
              'idade': [],
              'peso': [],
              'altura':[]}


nome = input('Digite seu nome: ')
print(f'Nome do aluno: {nome}')
idade = tratar_idade('Digite sua idade: ')
print(f'Idade : {idade} anos')
altura = tratar_altura('Digite sua altura: ')
print(f'Altura: {altura}')
peso = tratar_peso('Digite seu peso')
print(f'Peso: {peso}Kg')
resultado = calculo_imc(peso,altura)




if resultado <= 18.0:
    print(f'Abaixo do peso - imc de {resultado}% ')
elif resultado >= 18.5 and resultado <= 24.9:
    print(f'Peso normal - imc de {resultado}% ')
elif resultado >= 25.0 and resultado <= 29.9:
    print (f'Sobrepeso - imc de {resultado}% ')
else:
    print(f'Obesidade - imc de {resultado}% ')
```
