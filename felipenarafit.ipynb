```python
def tratar_nome(nome):
   while True:
        nome = input((nome))
        print('Nome cadastrado')
        return nome



    


def tratar_altura(funcao):
    while True:
        try:
            valor = float(input(funcao))
            print('Altura cadastrada')
            break
        except ValueError: 
            print('Digite um número válido')
    return valor
    

def tratar_peso(peso):
    while True:
        try:
            a1 = float(input(peso))
            print('Peso cadastrado')
            break
        except ValueError: 
            print('Digite um número válido')
    return a1


def tratar_idade(codigo):
    while True:
        try: 
         valor = int(input(codigo))
         print('Idade cadastrada')
         break
        except ValueError:
            print('Digite um número válido')
    return valor

def calculo_imc(peso,altura):
    resultado = peso / (altura**2)
    return resultado


```

```python
dicionario = {'nome': [],
              'idade': [],
              'peso': [],
              'altura':[],
              'resultado': []}


while True:
  nome = tratar_nome('Digite seu nome: ')
  print(f'Nome do aluno: {nome}')
  idade = tratar_idade('Digite sua idade: ')
  print(f'Idade : {idade} anos')
  altura = tratar_altura('Digite sua altura: ')
  print(f'Altura: {altura}')
  peso = tratar_peso('Digite seu peso')
  print(f'Peso: {peso}Kg')
  resultado = calculo_imc(peso,altura)


  dicionario['nome'].append(nome)
  dicionario['idade'].append(idade)
  dicionario['altura'].append(altura)
  dicionario['peso'].append(peso)
  dicionario['resultado'].append(resultado)
  break

for i in range(len(dicionario['nome'])):
    print(f'{dicionario['nome'][i]}, seu imc é de {round(dicionario['resultado'][i]*100,2)} %')




if resultado <= 18.0:
    print(f'Diagnóstico: Abaixo do peso - imc de {resultado:.2f}% ')
elif resultado >= 18.5 and resultado <= 24.9:
    print(f'Diagnóstico: Peso normal - imc de {resultado:.2f}% ')
elif resultado >= 25.0 and resultado <= 29.9:
    print (f'Diagnóstico: Sobrepeso - imc de {resultado:.2f}% ')
else:
    print(f'Diagnóstico: Obesidade - imc de {resultado:.2f}% ')
print(dicionario)
```
